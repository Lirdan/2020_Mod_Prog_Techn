####  ДИАГРАММЫ

```python
s = ['one','two','three ','four' ,'five']
x = [1, 2, 3, 4, 5]
z = np.random.random(100)
z1 = [10, 17, 24, 16, 22]
z2 = [12, 14, 21, 13, 17]

# СТОЛБИКИ
fig = plt.figure()
plt.bar(x, z1)
plt.title('Стовпчаста')
plt.grid(True)   # линии вспомогательной сетки

# ГИСТОГРММА
fig = plt.figure()
plt.hist(z)
plt.title('Гістограма')
plt.grid(True)

# КРУГОВАЯ
fig = plt.figure()
#plt.pie(x, labels=s)
plt.pie(x, [.1, .1, .1, .1, .1], labels = s, labeldistance = .5)
plt.title('Кугова')

# ОШИБКИ
fig = plt.figure()
plt.errorbar(x, z1, xerr=1, yerr=0.5)
plt.title('Похібки')
plt.grid(True)

plt.show()
```

####  КОНТУРНАЯ ДИАГРАММА

```python
# СОЗДАНИЕ МАТРИЦЫ ЗНАЧЕНИЙ
dat = np.random.random(200).reshape(20,10)

fig = plt.figure()
cr = plt.contour(dat) 
plt.title('Контур')

fig = plt.figure()
cf = plt.contourf(dat) 
plt.colorbar(cf)
plt.title('Конутр із колорбаром')

plt.show()
```
