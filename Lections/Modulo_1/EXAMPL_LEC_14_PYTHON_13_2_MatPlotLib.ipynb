```python
# ИМПОРТ МОДУЛЯ
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

print("Версия matplotlib : ", mpl.__version__)
print("Версия numpy : ", np.__version__)
```

### ЛИНЕЙНЫЕ ГРАФИКИ

```python
x = np.linspace(0, 2, 100)

# ИСПОЛЬЗУЕМ ЭКЗЕМПЛЯР ОБЪЕКТА FIGURE
fig, ax = plt.subplots()  # Создаем фигуру и область построения с осями 
ax.plot(x, x, label='линейная')  # Вывод графика на область 
ax.plot(x, x**2, label='квадратичная')  # Вывод еще одного графика 
ax.plot(x, x**3, label='кубическая')  # ... и еще одного 
ax.set_xlabel('аргумент')  # Добавляем обозначение X-оси
ax.set_ylabel('функция')  # Добавляем обозначение Y-оси
ax.set_title("ЗАВИСИМОСТИ")  # Добавляем название
ax.legend()  # Добавляем легенду
ax.grid(True)
```

### ДВА ГРАФИКА

```python
t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2*np.pi*t)
s2 = np.cos(4*np.pi*t)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
ax = axes.ravel()


line_1, = ax[0].plot(s1)
line_1.set_dashes([2, 2, 10, 2])  # 2pt line, 2pt break, 10pt line, 2pt break
ax[0].set_title("SIN")

# Using plot(..., dashes=...) to set the dashing when creating a line
line_2, = ax[1].plot(s2, dashes=[6, 2], label='Using the dashes parameter')
ax[1].set_title("COS")
ax[1].grid(True)

plt.show()
```

####  ВСТРОЕННЫЕ ГРАФИКИ (ИСПОЛЬЗОВАНИЕ GRIDSPEC)

```python
def plot_1(ax):
    ax.plot(s1)
    ax.set_xlabel('Angle', fontsize=12)
    ax.set_ylabel('Sin', fontsize=12)
    ax.set_title('SIN', fontsize=14)


def plot_2(ax):
    ax.plot(s2)
    ax.set_xlabel('Angle', fontsize=12)
    ax.set_ylabel('Cos', fontsize=12)
    ax.set_title('COS', fontsize=14)


def plot_3(ax):
    y = np.linspace(0, 2, 100)
    ax.plot(y, y**2)
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('x*x', fontsize=12)
    ax.set_title('QUADR', fontsize=14)

def plot_4(ax):
    y = np.linspace(0, 3, 10)
    ax.plot(y, np.exp(y))
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('e^x', fontsize=12)
    ax.set_title('EXPONENT', fontsize=14)

    
    
fig = plt.figure(constrained_layout=True, figsize=(12, 6))
import matplotlib.gridspec as gridspec

gs0 = gridspec.GridSpec(1, 2, figure=fig)
gs1 = gridspec.GridSpecFromSubplotSpec(3, 1, subplot_spec=gs0[0])
ax = fig.add_subplot(gs1[0])
ax.grid(True)
plot_1(ax)
ax = fig.add_subplot(gs1[1])
plot_2(ax)
ax = fig.add_subplot(gs1[2])
plot_3(ax)


gs2 = gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec=gs0[1])
ax = fig.add_subplot(gs2[0])
ax.grid(True)
plot_4(ax)



plt.show()
```
