```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
x = np.linspace(-50, 50, 100)
fig=plt.figure()
ax= fig.add_axes([0.1,0.1,0.9,0.9])
ax.plot(x, 2*(x**2)+3, color="#8800BB", ls="--",label="2x^2+3",alpha=0.5)
ax.set_ylim([0, 100])
ax.set_xlim([-50, 50])
ax.legend(loc=3)
ax.set_title("Графік функції")
ax.set_xlabel("X")
ax.set_ylabel("Y")
fig.savefig("MyFile.png",dpi=800)
```
